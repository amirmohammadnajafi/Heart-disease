## boosting

```r
library(gbm)
```

```r
set.seed(1)
boost <- gbm(target~ ., data =train,
distribution = "bernoulli", n.trees = 500,
interaction.depth = 3)
```

```r
yhat.boost <- predict(boost,
newdata = test, n.trees = 500,type="response")
```

```r

pred_class <- ifelse(yhat.boost > 0.5, 1, 0)
pred_class
```

```r
table(pred_class,test_y)
acc=mean(pred_class==test_y)
acc
```
